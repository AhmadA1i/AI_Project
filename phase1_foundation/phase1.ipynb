```python
class DataRecord:
    def __init__(self, record_id, features, label):
        self.record_id = record_id
        self.features = features
        self.label = label

    def display(self):
        print(f"Record ID: {self.record_id}")
        print("Features:")
        for feature_name, feature_value in self.features.items():
            if pd.notna(feature_value):
                print(f"  {feature_name}: {feature_value}")
        print(f"Label: {self.label}")
```

```python
symptom_cols = [col for col in df.columns if col != 'Disease']
records = []
for i in range(5):
    row = df.iloc[i]
    features = {col: row[col] for col in symptom_cols}
    record = DataRecord(record_id=row.name, features=features, label=row['Disease'])
    records.append(record)

# Display the first few records
for record in records:
    record.display()
```

```python
graph = {}

for _, row in df.iterrows():
    # Get all non-null symptoms in this row
    symptoms = [row[col] for col in symptom_cols if pd.notna(row[col])]
    
    # Connect every symptom to every other in same row
    for s in symptoms:
        if s not in graph:
            graph[s] = set()
        for other in symptoms:
            if other != s:
                graph[s].add(other)

# Convert sets to lists for display
graph = {k: list(v) for k, v in graph.items()}

# Print stats
total_nodes = len(graph)
total_edges = sum(len(v) for v in graph.values()) // 2

print(f"Total Nodes (unique symptoms): {total_nodes}")
print(f"Total Edges (co-occurrences) : {total_edges}")
```
