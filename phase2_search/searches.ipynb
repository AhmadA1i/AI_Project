```python
import pandas as pd
df = pd.read_csv('D:\AI_Project\data\dataset.csv')
# yaha par apna path add karna jasay agar desktop par ha to for example "desktop/AI_Project/data/dataset.csv"
```

```python
def get_symptom_columns(df):
    return [col for col in df.columns if col.startswith('Symptom')]

def build_graph(df):
    symptom_cols = get_symptom_columns(df)
    graph = {}
    for _, row in df.iterrows():
        symptoms = [row[col] for col in symptom_cols if pd.notna(row[col])]
        for s in symptoms:
            if s not in graph:
                graph[s] = set()
            for other in symptoms:
                if other != s:
                    graph[s].add(other)
    return {k: list(v) for k, v in graph.items()}

graph = build_graph(df)
```
